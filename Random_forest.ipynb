```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Mount Google Drive to access files
from google.colab import files
upload = files.upload()
df = pd.read_csv('Breast_cancer_detection_dataset.csv')
df.head(5)

# Drop empty columns
df = df.dropna(axis=1)

# Encode categorical data
from sklearn.preprocessing import LabelEncoder
labelencoder_Y = LabelEncoder()
df.iloc[:, 1] = labelencoder_Y.fit_transform(df.iloc[:, 1].values)

# Split the dataset into independent (X) and dependent (Y) data sets
X = df.iloc[:, 2:31].values
Y = df.iloc[:, 1].values

# Split the dataset into 25% testing and 75% training
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=0)

# Scale the data (feature scaling)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)

# Create a function for Random Forest model
def random_forest_model(X_train, Y_train):
    # Random Forest classifier
    rf_classifier = RandomForestClassifier(n_estimators=100, random_state=0)
    rf_classifier.fit(X_train, Y_train)

    # Print the model accuracy on the training data
    print('[Random Forest] Training Accuracy:', rf_classifier.score(X_train, Y_train))

    return rf_classifier

# Get the Random Forest model
rf_model = random_forest_model(X_train, Y_train)

# Make predictions on the test set
y_pred_rf = rf_model.predict(X_test)

# Calculate confusion matrix
cm_rf = confusion_matrix(Y_test, y_pred_rf)

# Print the confusion matrix
print("Random Forest Confusion Matrix:")
print(cm_rf)

# Calculate and print testing accuracy
accuracy_rf = accuracy_score(Y_test, y_pred_rf)
print('Random Forest Testing Accuracy =', accuracy_rf)

# Print the classification report
print("Random Forest Classification Report:")
print(classification_report(Y_test, y_pred_rf))
```
